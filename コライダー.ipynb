# コライダーをコントロールすることによって発生する擬似相関について

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num = 200
x = np.random.randn(num)
y = np.random.randn(num)
plt.scatter(x,y)
np.corrcoef(x,y)
```

```python
z = x + y

new_x = []
new_y = []

for i in range(num):
    if z[i] > 1:
        new_x.append(x[i])
        new_y.append(y[i])
```

```python
plt.scatter(new_x,new_y)
np.corrcoef(new_x,new_y)
```

zを統制することで、擬似相関が生まれている。
例で考える
ある会社の採用面接では人柄と技術の２つを採点して、ある合格点以上で採用することとする。面接に受けに来る人の能力は独立だが、採用基準を設けることによって、合格者の集団にバイアスが生じる。合格者だけを見ると、人柄と技術には負の相関がある。これがコライダーと呼ばれるものの正体。

# 統計的因果探索

```python
x1 = np.random.normal(0,1,1000)
x2 = x1 * 0.8 + np.random.normal(0,1,1000)
plt.xlabel('x1', fontsize = 16)
plt.ylabel('x2', fontsize = 16)
plt.scatter(x1,x2)
```

```python
x2 = np.random.normal(0,1,1000)
x1 = x2 * 0.8 + np.random.normal(0,1,1000)
plt.xlabel('x1', fontsize = 16)
plt.ylabel('x2', fontsize = 16)
plt.scatter(x1,x2)
```
